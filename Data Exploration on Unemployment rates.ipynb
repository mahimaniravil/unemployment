## FILTER THE DATA

```python
#Show rows having estimated employed greater than or equal to 100000 ,estimated unemployment rate less than 60 in various region starts with'A
import pandas as pd
display(df.loc[(df['Estimated_Employed']>=100000) & (df['Estimated_Unemployment_Rate']< 60) & (df['Region'].str.startswith('A'))])
```

```python
#using Numpy
import pandas as pd 
import numpy as np
filtered_values = np.where((df['Estimated_Employed']>=100000) & (df['Estimated_Unemployment_Rate']< 60) & (df['Region'].str.startswith('A')))
print(filtered_values)
display(df.loc[filtered_values])
```

```python
#show rows having estimated employed greater than 2000000 and estimated unemployment rate <50 in region that starts with 'k'
import pandas as pd
display(df.loc[(df['Estimated_Employed']>1000000) & (df['Estimated_Unemployment_Rate']< 10) & (df['Region'].str.startswith('K'))])
```

```python
#show the rows having estimated labour participation rate less than 40.
import pandas as pd
display(df.loc[(df['Estimated_Labour_Participation_Rate']<40)])
```

```python
import pandas as pd
display(df.loc[(df['Estimated_Unemployment_Rate']<10) & (df['Estimated_Employed']<1000000) & (df['Estimated_Labour_Participation_Rate']<25)])
```

## VISUALISE THE DATA

```python
#show a bargraph showing the unemployment rates in different region
import matplotlib.pyplot as plt
import pandas as pd
df.Region.value_counts().nlargest(30).plot(kind='bar', figsize=(10,5),color='red')
plt.title("unemployment rates in different region")
plt.ylabel('unemployment rates')
plt.xlabel('Region');
plt.show()
```

```python
#Show a line graph of labour participation rate in different region
import matplotlib.pyplot as plt
plt.plot(df['Region'],df['Estimated_Labour_Participation_Rate'],color='green')
plt.xlim(0,28)
plt.xlabel('Region',color='red')
plt.ylabel('Estimated_Labour_Participation_Rate',color='red')
plt.title('Estimated Labour Participation Rate in India over the Region',color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Region'], bins=28, color='purple')
plt.xlabel('Region')
plt.ylabel('Estimated_Employed')
plt.title('Histogram showing Estimated Employed in different Region')
plt.show()
```

```python
 
import seaborn as sns
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Estimated_Labour_Participation_Rate', y='Estimated_Unemployment_Rate')
plt.xlabel('Region')
plt.ylabel('Estimated_Unemployment_Rate')
plt.title('Box Plot of Estimated_Unemployment_Rate')
plt.show()
```
